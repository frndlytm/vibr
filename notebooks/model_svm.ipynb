```python
import numpy as np

features = np.load("data/tp_source_trimmed.npy", allow_pickle=True)
moods = np.load("data/moods_target_trimmed.npy", allow_pickle=True)[:, 1]


m_len = len(moods[0])

for i in range(len(moods)):
    moods[i] = np.array(moods[i])

moods = np.stack(moods)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

class MultiClassSVM:

    def __init__(self):
        self.fit_models = None

    def fit(self, X, y):
        classes = len(y[0])
        self.fit_models = []
        for i in range(classes):
            print("feature ", i, " of ", classes, "...")
            m_i = SVC(C=1, kernel='rbf')
            m_i.fit(X,  y[:,i])
            self.fit_models.append(m_i)

        print(len(self.fit_models))

    def predict(self, x_ex):
        out = []
        for ex in x_ex:
            classes = []
            for i in range(len(self.fit_models)):
                classes.append(self.fit_models[i].predict([ex])[0])
            out.append(classes)

        return out

    def feature_accuracy(self, x_ex, y_known):
        feats = {}
        for f in range(len(y_known[0])):
            true = y_known[:,f]
            pred = self.fit_models[f].predict(x_ex)
            feats[f] = accuracy_score(true, pred), precision_score(true, pred), recall_score(true, pred)

        return feats
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

features = preprocessing.normalize(features)

X_train, X_test, y_train, y_test = train_test_split(features, moods, test_size=0.2, shuffle=True)
```

```python
multi_model = MultiClassSVM()
multi_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

stats = multi_model.feature_accuracy(X_test, y_test)
```

```python
import json

with open('svm_results.json', 'w') as fp:
    json.dump(stats, fp, indent=4)

```
